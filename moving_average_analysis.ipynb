### Reading Dataset

```python
import pandas as pd
import numpy as np
```

```python
df_pr = pd.read_csv('OPENAIERC-USD.csv', na_values=['null'], index_col='Date', parse_dates=True, infer_datetime_format=True)
```

### Simple Moving Averages

```python
## simple moving average

df_pr['Open:10 days rolling'] = df_pr['Open'].rolling(window=10, min_periods=1).mean()
```

```python
df_pr[['Open', 'Open:10 days rolling']].plot(figsize = (12,6))
```

```python
df_pr[['Open', 'Open:10 days rolling']].plot(xlim = ['2022-12-07', '2023-02-28'],figsize = (12,6))
```

```python
df_pr['Open:30 days rolling'] = df_pr['Open'].rolling(window=30, min_periods=1).mean()
df_pr['Open:50 days rolling'] = df_pr['Open'].rolling(window=50, min_periods=1).mean()
```

```python
df_pr[['Open', 'Open:10 days rolling','Open:30 days rolling','Open:50 days rolling' ]].plot(figsize = (12,6))
```

```python
df_pr[['Open', 'Open:10 days rolling','Open:30 days rolling','Open:50 days rolling' ]].plot(xlim = ['2022-12-07', '2023-02-28'],figsize = (12,6))
```

###  Cummulative moving average

```python
df_pr['Open'].expanding().mean().plot(figsize = (12,6))
```

### Exponential Moving Average

```python
# alpha = smoothening factor 
df_pr['ema_open_0.1'] = df_pr['Open'].ewm(alpha =0.1, adjust = False).mean()
```

```python
df_pr[['Open', 'Open:50 days rolling', 'ema_open_0.1'  ]].plot(xlim = ['2022-12-07', '2023-02-28'],figsize = (12,6))
```

```python
#alpha = 0.3
df_pr['ema_open_0.3'] = df_pr['Open'].ewm(alpha =0.3, adjust = False).mean()
```

```python
df_pr[['Open', 'Open:50 days rolling', 'ema_open_0.1', 'ema_open_0.3' ]].plot(xlim = ['2022-12-07', '2023-02-28'],figsize = (12,6))
```

### Exponential Weighted Moving Average

```python
## alpha = 2/(span+1)
## span = duration 
## where, alpha = smoothening factor 

df_pr['ewma_10days'] = df_pr['Open'].ewm(span = 10, adjust = False).mean()
```

```python
df_pr[['Open', 'ema_open_0.3' , 'ewma_10days']].plot(xlim = ['2022-12-07', '2023-02-28'], figsize = (12,6))
```

```python
df_pr[['Open', 'ema_open_0.3' , 'ewma_10days']].plot( figsize = (12,6))
```
